```python
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd

url = "https://www.rottentomatoes.com/top/bestofrt/?year=2019"
res = req.urlopen(url)

# HTML 분석
soup = BeautifulSoup(res, "html.parser")

topMovie2019 = soup.select("#top_movies_main > div > table a")
topMovie2019_Rating = soup.select("#top_movies_main > div > table span.tMeterScore")
topMovie2019_Reviews = soup.select("#top_movies_main > div > table td.hidden-xs")
  
topMovie2019Lists = [] # 영화제목
countLists = [] # 랭킹
Rating = [] # 예매율
Reviews = [] # 리뷰 갯수

count = 1
for movie100 in topMovie2019:
    #print(count, movie100.string.strip().replace("(2019)",""))
    countLists.append(count)
    topMovie2019Lists.append(movie100.string.strip().replace("(2019)",""))
    count += 1
    
for rate in topMovie2019_Rating:
    Rating.append(rate.string)

for review in topMovie2019_Reviews:
    Reviews.append(review.string)
```
